```python
def churn_stats_by_gender(df):
    # Total customers per gender
    total_customers = df['gender'].value_counts()
    
    # Total churners per gender
    total_churners = df[df['Churn_Binary'] == 1]['gender'].value_counts()
    
    # Churn percentage
    churn_pct = (total_churners / total_customers) * 100
    
    return pd.DataFrame({
        'Total Customers': total_customers,
        'Total Churners': total_churners,
        'Churn Percentage (%)': churn_pct
    })

# Example usage:
stats = churn_stats_by_gender(df)
print(stats)
```

```python
df["tenure"].sort_values(ascending=False).unique()
```

```python
#Churn rate by tenure
#Bin tenure into 5 segments
bins = [0, 3, 6, 12, 24, 48, 72]
labels = ['0-3mo', '3-6mo', '6-12mo', '1-2y', '2-4y', '4-6y']
df['Tenure_Group'] = pd.cut(df["tenure"], bins=bins, labels=labels, include_lowest=True)
```

```python
#Calculate churn rate by Tenure Group
Tenure_churn = df.groupby('Tenure_Group')['Churn_Binary'].mean() * 100
print(Tenure_churn)
```

```python
plt.figure(figsize=(8, 4))
Tenure_churn.plot(kind='bar', color=['red', 'orange', 'green', 'yellow','blue','pink'])
plt.title('Churn Rate by Tenure Group', fontsize=14)
plt.xlabel('Tenure Type', fontsize=12)
plt.ylabel('Churn Rate (%)', fontsize=12)
plt.xticks(rotation=0)  # Horizontal x-axis labels
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show();
```

```python
#Customers with a 0-3 month tenure have the highest churn rate which is 56.2 percent. Customers who have a tenure of more than 4 years have the lowest 
#churn rate, which seems to be 9.5 percent according to our data.
```

```python
#Filtering the customers who have no internet, since TechSupport is not relevant for them.
df_internet = df[df['TechSupport'] != 'No internet service']
```

```python
tech_support_churn = df_internet.groupby('TechSupport')['Churn_Binary'].mean() * 100
print(tech_support_churn)
```

```python
plt.figure(figsize=(8, 4))
tech_support_churn.plot(kind='bar', color=['red', 'orange'])
plt.title('Churn Rate by Tech Support', fontsize=14)
plt.xlabel('Tech Support', fontsize=12)
plt.ylabel('Churn Rate (%)', fontsize=12)
plt.xticks(rotation=0)  # Horizontal x-axis labels
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show();
```

```python
from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(
    df_internet['TechSupport'], 
    df_internet['Churn_Binary']
)
chi2, p_value, _, _ = chi2_contingency(contingency_table)
chi2, p_value
```

```python
#p-value is <0.05, there is enough evidence for a significant association between receiving technical support and churn rate. 
#Customers who receive technical support are 2.7x less likely to churn suggesting an effective method for retaining customers. 
```

```python
sns.boxplot(data=df_internet, x='TechSupport', y='MonthlyCharges', hue='Churn',
dodge=True,  
width=0.6,
gap=0.3,
palette={'Yes': 'salmon', 'No': 'green'})
plt.title("Monthly Charges by Tech Support and Churn Status")
plt.show();
```

```python
#Median charge is lower for non-churned customers → suggests cheaper plans retain users.
#If the customer doesn't receive tech support, the interquartile range (IQR) is wider for non-churned customers, indicating 
#higher variability in monthly charges among retained customers.
#Higher median for churned customers meaning churned customers still pay slightly more even with support, high costs push users away.
#Insight: Customers with No Tech Support + High Charges are most likely to churn.-> Strategy: Include tech support in high-cost plans to mitigate churns. 
```

```python
#How do service bundles impact churn?
grouped = df.groupby('InternetService')['Churn_Binary'].mean()
print(grouped)
```

```python
churn_by_bundle = df.groupby(['InternetService', 'PhoneService'])['Churn_Binary'].mean().reset_index()  
print(churn_by_bundle.sort_values('Churn_Binary', ascending=False))
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(
    data=churn_by_bundle,
    x='InternetService',
    y='Churn_Binary',
    hue='PhoneService',
    palette='Blues',
    order=['Fiber optic', 'DSL', 'No']  # Optional: Force order
)
plt.title('Churn Rate by Internet + Phone Service Bundle', pad=20)
plt.xlabel('Internet Service')
plt.ylabel('Churn Rate (0-1)')
plt.ylim(0, 1)  # Adjust if churn is a %
plt.legend(title='Phone Service', loc='upper right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
sns.despine()
plt.show()
```

```python
#Insights: Highest churn rate is observed for customers who use both fiber optic and phone services. (Likely due to high expectations/high competition in this segment)
#Second Highest: DSL + No phone service, (Older technology users may be frustrated with speeds)
#Third Highest: DSL + Phone service 
#Lowest churn rate among customers who use only phone service. These customers are likely less tech-dependent.
```

```python
df.groupby(['InternetService','PhoneService'])['MonthlyCharges'].mean().plot(kind='bar')
plt.xticks(rotation=0)  # Horizontal x-axis labels
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xlabel('Service Bundle(Internet, Phone)', fontsize=12)
plt.ylabel('Average Monthly Charges', fontsize=12)
plt.title('Average Monthly Charges per Service Bundle', fontsize=14)
plt.show()
```
